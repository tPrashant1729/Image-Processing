```python
import cv2 as cv
import numpy as np

# Replace 'path_to_image.jpg' with the path to your image file
img = cv.imread('D:\Dnk_project\output\chelsea.png')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# Function to change contrast and brightness of the image

def adjust_contrast_brightness(image, alpha, beta):
    # Apply the linear transformation: new_pixel_value = alpha * original_pixel_value + beta
    new_image = np.clip(alpha * image + beta, 0, 255).astype(np.uint8)
    return new_image

# Adjust the contrast and brightness
contrast_factor = 1.5  # Increase for higher contrast, decrease for lower contrast
brightness_factor = 50  # Increase to make the image brighter, decrease to make it darker

adjusted_image = adjust_contrast_brightness(img_rgb, contrast_factor, brightness_factor)

# Display the original and adjusted images side by side
combined_image = np.hstack((img_rgb, adjusted_image))
plt.imshow( combined_image)
plt.axis('off')
plt.show()
```
